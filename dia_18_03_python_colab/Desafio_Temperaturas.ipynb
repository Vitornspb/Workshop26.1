```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

temperaturas = np.random.uniform(20, 35, 7)
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]

df = pd.DataFrame({
    'Dia': dias,
    'Temperatura': temperaturas
})

media = df['Temperatura'].mean()
maxima = df['Temperatura'].max()
minima = df['Temperatura'].min()

print(f"Média: {media:.2f}")
print(f"Máxima: {maxima:.2f}")
print(f"Mínima: {minima:.2f}")

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(df['Dia'], df['Temperatura'], marker='o', color='blue')
plt.title('Variação da Temperatura ao Longo da Semana')
plt.xlabel('Dias')
plt.ylabel('Temperatura (°C)')
plt.xticks(rotation=45)
plt.axhline(media, color='green', linestyle='--', label='Média')
plt.legend()

plt.subplot(1, 3, 2)
plt.bar(df['Dia'], df['Temperatura'], color='skyblue')
plt.title('Temperatura por Dia')
plt.xlabel('Dias')
plt.ylabel('Temperatura (°C)')
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
plt.scatter(df['Dia'], df['Temperatura'], color='red')
plt.title('Dispersão de Temperatura')
plt.xlabel('Dias')
plt.ylabel('Temperatura (°C)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
